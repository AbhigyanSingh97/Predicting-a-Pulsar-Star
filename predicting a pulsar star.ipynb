```r
smp_size <- floor(nrow(data) * 0.75)
train_data <- sample(seq_len(nrow(data)),size = smp_size)
set.seed(123)
train <- data[train_data,]
test <- data[-train_data,]
```

```r
nrow(data)
nrow(train)
nrow(test)
```

```r
model <- glm(target_class ~ . , data = train,family = binomial())
```

```r
summary(model)
```

```r
plot(model)
```

```r
prediction <- predict(model,test,type='response')
```

```r
test$prediction<- ifelse(prediction >= 0.5,1,0)
```

```r
prediction
```

```r
xtabs(~target_class + prediction,test)
```

```r
accuracy(test$target_class,test$prediction,0.5)
```
